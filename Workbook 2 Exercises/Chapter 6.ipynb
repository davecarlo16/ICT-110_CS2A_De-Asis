```python
Given the code below

text = 'Business Analytics'

What will the following expressions produce?
```

```python
text = 'Business Analytics'
print(text[4])  # n
print(text[2:4]) # si
print(text[4:]) # ness Analytics
print(text[:4]) # Busi
print(text[-2]) # c
print(text[-4:-2]) # ti
print(text[-2:]) # cs
print(text[:-2]) # Business Analyti
```
