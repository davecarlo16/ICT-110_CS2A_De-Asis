####Exercise 1
Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

```python
def compute_pay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        pay = (40 * rate) + ((hours - 40) * 1.5 * rate)
    return pay

hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

total_pay = compute_pay(hours, rate)
print("Pay:", total_pay)
```

####Challenge
Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

```python
Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

Run the program repeatedly to test the various different values for input.
```

```python
def computegrade(score):
    if 97 <= score <= 100:
        return '1.0'
    elif 93 <= score < 97:
        return '1.25'
    elif 89 <= score < 93:
        return '1.5'
    elif 85 <= score < 89:
        return '1.75'
    elif 81 <= score < 85:
        return '2.0'
    elif 77 <= score < 81:
        return '2.25'
    elif 74 <= score < 77:
        return '2.5'
    elif 0 <= score < 74:
        return '3.0'
    else:
        return 'Invalid score entered'

score = float(input("Enter the score: "))
grade = computegrade(score)
print("Grade:", grade)
```
