####Exercise2
Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours.

```python
hours_worked = float(input("Enter hours worked: "))
hourly_rate = float(input("Enter hourly rate: "))

if hours_worked <= 40:
    salary = hours_worked * hourly_rate
else:
    salary = (40 * hourly_rate) + ((hours_worked - 40) * 1.5 * hourly_rate)

print("Salary:", salary)
```

####Challenge
Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

```python
score = float(input("Enter the score (between 0 and 100): "))

if 97 <= score <= 100:
    grade = 1.0
elif 93 <= score < 97:
    grade = 1.25
elif 89 <= score < 93:
    grade = 1.5
elif 85 <= score < 89:
    grade = 1.75
elif 81 <= score < 85:
    grade = 2.0
elif 77 <= score < 81:
    grade = 2.25
elif 74 <= score < 77:
    grade = 2.5
elif 0 <= score < 74:
    grade = 3.0
else:
    grade = None

if grade is not None:
    print("Grade:", grade)
else:
    print("Invalid score entered.")
```
