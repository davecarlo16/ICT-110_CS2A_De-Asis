####Exercise 1
Investigate on the difference between is and == given the following variables.

```python
a = 2
b = 2.0
c = a
```

```python
print(a == b)  # True, because the values are equal (2 == 2.0)
print(a is b)  # False, because they are different objects with different memory locations
print(a == c)  # True, because the values are equal (2 == 2)
print(a is c)  # True, because they are the same object (both referring to the integer value 2)
```

###Challenge 1
Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

```python
Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333
```

```python
#Edited code
total = 0
count = 0

while True:
    num = input("Enter a number: ")
    if num == 'done':
        break
    try:
        num = float(num)
        total += num
        count += 1
    except ValueError:
        print("Invalid input")
        continue

if count == 0:
    print("No numbers entered")
else:
    average = total / count
    print(total, count, average)
```

##Challenge 2
Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
numbers = []

while True:
    num = input("Enter a number: ")
    if num == 'done':
        break
    try:
        num = float(num)
        numbers.append(num)
    except ValueError:
        print("Invalid input")
        continue

if numbers:
    print("Maximum:", max(numbers))
    print("Minimum:", min(numbers))
else:
    print("No numbers entered")
```
